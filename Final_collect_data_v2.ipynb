```python

#Step 16 Check work flow OK, try to use on all 10yr data

def load_standings(path, conference_label, year):
    raw = pd.read_csv(path, header=None)
    s = raw.iloc[:, 0].astype(str).str.replace('"', '', regex=False)
    parts = s.str.split('\t', expand=True)
    
    header = parts.iloc[0]
    df = parts.iloc[1:].reset_index(drop=True)
    df.columns = header
    
    first_col = df.columns[0]
    df = df.rename(columns={first_col: "Team"})
    
    df["Team_clean"] = df["Team"].astype(str).str.replace("*", "", regex=False).str.strip()
    df["Conference"] = conference_label
    df["Year"] = year
    
    df = df[["Team", "Team_clean", "W", "L", "Conference", "Year"]]
    return df
```

```python
#Step 17 Use new fun to load

east_2024 = load_standings("NBA_Standings_East_2023_2024.csv", "East", 2024)
west_2024 = load_standings("NBA_Standings_West_2023_2024.csv", "West", 2024)

print(east_2024.head())
print(east_2024.shape)

print(west_2024.head())
print(west_2024.shape)
```

```python

#Step 18 concat

standings_2024 = pd.concat([east_2024, west_2024], ignore_index=True)

print(standings_2024.head())
print(standings_2024.shape)
print(standings_2024["Conference"].value_counts())
```

```python

#Step 19 Also run stats

stats_2024 = pd.read_csv("NBA_Stats_2023_2024.csv")
stats_2024 = stats_2024[~stats_2024["Team"].astype(str).str.contains("League Average", na=False)]
stats_2024["Team_clean"] = stats_2024["Team"].astype(str).str.replace("*", "", regex=False).str.strip()
stats_2024["Year"] = 2024

print(stats_2024[["Team", "Team_clean", "Year"]].head())
print(stats_2024.shape)
print("unique teams in stats_2024:", stats_2024["Team_clean"].nunique())
```

```python

#Step 20 also merge

merged_2024_auto = pd.merge(
    stats_2024,
    standings_2024,
    on=["Team_clean", "Year"],
    how="inner"
)

merged_2024_auto = merged_2024_auto.rename(columns={"Team_x": "Team"}).drop(columns=["Team_y"])

print(merged_2024_auto[["Team", "W", "L", "Conference"]].head())
print(merged_2024_auto.shape)
print("unique teams:", merged_2024_auto["Team_clean"].nunique())
```

```python
# Step 21 check 2 method equal
print(merged_2024_auto.equals(merged))
print(merged_2024_auto.head())
print(merged_2024_auto.shape)
```

```python

# Step 22 make path name

def make_paths(year):
    prev = year - 1
    stats_path = f"NBA_Stats_{prev}_{year}.csv"
    east_path = f"NBA_Standings_East_{prev}_{year}.csv"
    west_path = f"NBA_Standings_West_{prev}_{year}.csv"
    return stats_path, east_path, west_path
```

```python

# Step 23 check naming ok

stats_path_2024, east_path_2024, west_path_2024 = make_paths(2024)

print(stats_path_2024)
print(east_path_2024)
print(west_path_2024)
```

```python

#Step 24 auto do upward steps

def build_season(year):
    stats_path, east_path, west_path = make_paths(year)
    
    stats = pd.read_csv(stats_path)
    stats = stats[~stats["Team"].astype(str).str.contains("League Average", na=False)]
    stats["Team_clean"] = stats["Team"].astype(str).str.replace("*", "", regex=False).str.strip()
    stats["Year"] = year
    
    east = load_standings(east_path, "East", year)
    west = load_standings(west_path, "West", year)
    standings = pd.concat([east, west], ignore_index=True)
    
    merged = pd.merge(
        stats,
        standings,
        on=["Team_clean", "Year"],
        how="inner"
    )
    
    merged = merged.rename(columns={"Team_x": "Team"}).drop(columns=["Team_y"])
    
    return merged
```

```python

#Step 25 check new fun build_season ok

season_2024 = build_season(2024)

print(season_2024[["Team", "W", "L", "Conference"]].head())
print(season_2024.shape)
print("unique teams:", season_2024["Team_clean"].nunique())
print(season_2024.equals(merged_2024_auto))
```

```python

#Step 26 use small amount of yr to check every thing ok

years = [2022, 2023, 2024]

all_seasons_list = []

for y in years:
    df_y = build_season(y)
    df_y["Season"] = f"{y-1}-{y}"
    all_seasons_list.append(df_y)

all_seasons = pd.concat(all_seasons_list, ignore_index=True)

print(all_seasons[["Season", "Team", "W", "L", "Conference"]].head(20))
print(all_seasons["Season"].value_counts())
print(all_seasons.shape)
```

```python

#Step 27 define use yr data

years = list(range(2016, 2026))  # 2016, 2017, ..., 2025
print(years)
```

```python

# Step 28 Build all

all_seasons_list = []

for y in years:
    print(f"Building season for year = {y} ...")
    df_y = build_season(y)          
    df_y["Season"] = f"{y-1}-{y}"   
    
    print("  shape:", df_y.shape)
    print("  unique teams:", df_y["Team_clean"].nunique())
    
    all_seasons_list.append(df_y)
```
